```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X = df.drop(columns=["Attrition"])  
y = df["Attrition"]  


X_scaled = scaler.fit_transform(X)


X = pd.DataFrame(X_scaled, columns=X.columns)

```

```python
X
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)

rfe = RFE(model, n_features_to_select=6)

rfe.fit(X, y)


selected_features = X.columns[rfe.support_]

print("Top 6 selected features:", list(selected_features))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_selected = X[selected_features].copy()
df_selected["Attrition"] = y  


corr_matrix = df_selected.corr()


plt.figure(figsize=(10, 6))


sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)


plt.title("Heatmap of Selected Features & Attrition")


plt.show()
```

```python
#Interpretation of the Heatmap --
#The analysis reveals that selected characteristics alongside attrition rates show these correlations during the study period.

# 1)OverTime & Attrition (+0.25, Weak Positive Correlation)
#Staff members who perform additional hours at work show increased chances of job departure.
#Indicates potential burnout or dissatisfaction.

# 2)JobInvolvement & Attrition (-0.13, Weak Negative Correlation)
#People with strong job involvement display a minimal decrease in job departure activity.
#Suggests engagement can reduce attrition.

# 3)MonthlyIncome & Attrition (-0.16, Weak Negative Correlation)
#Cash rewards seem to slightly delay the departure of employees.
#The compensation rates providers can help workers stay with the company.

# 4)YearsSinceLastPromotion & Attrition (-0.16, Weak Negative Correlation)
#Staff who have not gotten a promotion during recent times show slightly higher risk of departing the organization.
#The current data indicates that career stagnation can serve as a motivating factor in employee decision to leave.

# 5)YearsWithCurrManager & YearsInCurrentRole (+0.69, Strong Positive Correlation)
#Workers who maintain their current position also extend their time under the same leadership.
#Workers remain committed because their jobs provide stability along with restricted opportunities for moving within the organization.

#Final Takeaways:
#The combination of long working hours causes employees to leave their jobs whereas attractive compensation and employee involvement help retain them.
#Turnover incidence increases due to promotion delays even though this effect remains weak.
#Employees tend to remain with the same manager when they extend their time in a particular position.
```

```python
# Hypothesis Results -

## Introduction:
#This analysis evaluates the relationship between various independent variables and employee attrition using Ordinary Least Squares (OLS) regression. 
#The objective is to determine whether the hypotheses are supported or rejected based on statistical significance (p-values) and the direction of 
#influence (coefficients).  

#Hypothesis 1: JobInvolvement significantly affects attrition.
#Coefficient: -0.1956  
#P-value: 0.000 (Significant at p < 0.05)  
#Interpretation: Employees with higher job involvement are less likely to leave. A negative coefficient indicates higher engagement reduces attrition.  
#Conclusion: H1 is accepted.  

# Hypothesis 2: MonthlyIncome significantly impacts employee attrition.
#Coefficient: -0.000017  
#P-value: 0.827 (Not significant)  
#Interpretation: Monthly income does not have a significant effect on attrition. This suggests that salary alone does not determine whether 
#an employee stays or leaves.  
#Conclusion: H2 is rejected.  

# Hypothesis 3: OverTime is positively correlated with attrition.
#Coefficient: 0.1995  
#P-value: 0.000 (Significant at p < 0.05)  
#Interpretation: Employees who work overtime are more likely to leave. A positive coefficient confirms that excessive overtime can lead to burnout, 
#increasing attrition.  
#Conclusion: H3 is accepted.  

# Hypothesis 4: YearsInCurrentRole influences employee attrition.
#Coefficient: -0.2068  
#P-value: 0.000 (Significant at p < 0.05)  
#Interpretation: Employees who have been in their current role for longer tend to stay, suggesting that job stability reduces attrition.  
#Conclusion: H4 is accepted.  

# Hypothesis 5: YearsSinceLastPromotion affects attrition rates.
#Coefficient: 0.1966  
#P-value: 0.000 (Significant at p < 0.05)  
#Interpretation: Employees who have not been promoted for a long time are more likely to leave. The positive coefficient indicates that career 
#stagnation increases attrition.  
#Conclusion: H5 is accepted.  

#Hypothesis 6: YearsWithCurrManager influences employee attrition.
#Coefficient: -0.1346  
#P-value: 0.027 (Significant at p < 0.05)  
#Interpretation: Employees who stay longer with the same manager are less likely to leave. This suggests that a good relationship with a manager 
#enhances retention.  
#Conclusion: H6 is accepted.  


## Conclusion:
#The OLS regression results support five out of six hypotheses, indicating that job involvement, overtime, tenure in the current role, last promotion, 
#and manager relationships significantly influence attrition. However, monthly income does not have a statistically significant impact, suggesting 
#that salary alone is not a key factor in employee turnover.
```

```python
selected_features = ['JobInvolvement', 'MonthlyIncome', 'OverTime', 'YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager']
X_selected = X[selected_features]
vif_data = pd.DataFrame()
vif_data["Feature"] = X_selected.columns
vif_data["VIF"] = [variance_inflation_factor(X_selected.values, i) for i in range(X_selected.shape[1])]

print(vif_data)
```

# residual analysis of the data

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns

# Define features (x_rfe) and target (y_actual)
x_rfe = df[selected_features]  # Ensure df contains your dataset
y_actual = df["Attrition"]  # Target variable

# Initialize and fit the model on the full dataset
lm = LinearRegression()
lm.fit(x_rfe, y_actual)

# Predict attrition using the selected features
y_pred = lm.predict(x_rfe)

# Calculate residuals
res = y_pred - y_actual

# Plot residual distribution
fig = plt.figure()
sns.histplot(res, bins=20, kde=True)  # Replaced sns.distplot with sns.histplot
fig.suptitle("Error Terms", fontsize=20)
plt.xlabel("Residuals")
plt.ylabel("Frequency")
plt.show()
```

```python
#The residuals exhibit a bimodal distribution, indicating potential missing variables or distinct subgroups in the data.
#The errors are not normally distributed, which may violate regression assumptions and affect prediction reliability.
#The presence of skewness suggests the model may be biased in certain predictions, requiring transformations or additional features.
#Possible heteroscedasticity (unequal variance) might be present, which can be verified using a residual vs. fitted values plot.






```
